## End to end word Context

```python
ID_Vocab = "makan"
ID_Vocab = ID_Vocab.lower()
```

```python
#Search the div
r = requests.get("https://tr-ex.me/translation/indonesian-english/" + ID_Vocab)
c = r.content
from bs4 import BeautifulSoup

soup = BeautifulSoup(c)
main_content = soup.find_all('div', attrs = {'class': 'm'})
```

```python
import re
main_content = str(main_content)

main_content = main_content.partition('\n')[0]
Result = re.sub('<.*?>', '', main_content)

Result = Result.replace('"',"")
Result = Result.replace('”',"")
Result = Result.split("!|?|.")
```

```python
A = "".join(Result)
A = A.replace(']',"")
A = A.replace('[',"")
```

```python
A
```

```python
ListResult = list(A.split(".,"))
```

```python
ListResult
```

```python
#Result= [s.replace("[", "") for s in list_of_strings]
```

```python
Result_Context = []

for i in range(0, len(ListResult), 2):
 Result_Context.append([ListResult[i], ListResult[i+1]])
np.shape(Result_Context)
```

```python
Result_Context
```

```python
df = pd.DataFrame(Result_Context, columns =['ID', 'EN'])
df
```

```python
for item in df.EN:
  item = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", item)

df["ID"] = df.ID.str.replace('~[^a-zA-Z0-9\s]+~', '')
```
